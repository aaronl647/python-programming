# Grouping cryptocurrency growth rate over the past year

```python
#this line adds a date to Bitcoin dataset
bt_prices = pd.DataFrame(bt_hist)
bt_prices['date'] = pd.date_range(start='07/23/2018', periods=len(bt_prices), freq='D')
```

```python
eth_prices = pd.DataFrame(eth_hist)
eth_prices['date'] = pd.date_range(start='07/23/2018', periods=len(eth_prices), freq='D')
```

```python
xrp_prices = pd.DataFrame(xrp_hist)
xrp_prices['date'] = pd.date_range(start='07/23/2018', periods=len(xrp_prices), freq='D')
```

```python
lite_prices = pd.DataFrame(lite_hist)
lite_prices['date'] = pd.date_range(start='07/23/2018', periods=len(lite_prices), freq='D')
```

```python
btc_prices = pd.DataFrame(btc_hist)
btc_prices['date'] = pd.date_range(start='07/23/2018', periods=len(btc_prices), freq='D')
```

```python
###Merges the 'Bitcoin' and 'Ethereum' datasets together
b_e_prices = pd.merge(bt_prices, eth_prices, on='date')
```

```python
#Merges Ripple into the same dataset
b_e_x_prices = pd.merge(b_e_prices, xrp_prices, on='date')
```

```python
b_e_x_prices
```

```python
b_e_x_prices.rename(columns={'prices_x': 'Bitcoin',
                           'prices_y': 'Ethereum',
                           'prices':'XRP(Ripple)'}, inplace=True)
```

```python
#Merges Litecoin and Bitcoin-Cash datasets together
b_e_x_l_prices = pd.merge(b_e_x_prices, lite_prices, on='date')
coin_prices = pd.merge(b_e_x_l_prices, btc_prices, on='date')
```

```python
coin_prices
```

```python
coin_prices.rename(columns={'prices_x':'Litecoin',
                           'prices_y':'Bitcoin-Cash'},inplace=True)
```

# Map of all active 'Bitcoin' markets around the world

```python
#Shows how many active markets there are, by country
test_data = [dict(type='choropleth',
                 autocolorscale=False,
                 locations=currency_merge['Country'],
                 z=currency_merge['count'],
                 locationmode='country names',
                 marker=dict(
                     line=dict(
                         color='rgb(255,255,255)',
                         width=1
                     )),
                 colorbar=dict(
                     title='Number of Markets'),
                  colorscale = 'Portland'
                 )] 
layout = dict(
            title='Active \'Bitcoin\' Markets',
            geo = dict(
                scope='world',
                lakecolor='rgb(255,255,255)'),
            )

fig = dict(data=test_data, layout=layout)

pyo.iplot(fig,filename='d3-cloropleth-map')
```
